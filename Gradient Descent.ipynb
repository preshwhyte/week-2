#  Gradient descent as a function

```python
def Grad_desc(derivative_func, initial_guess, multiplier, precision, max_inter):
    new_x=initial_guess
    
    
   

    x_list=[new_x]
    slope_list=[derivative_func(new_x)]

    for n in range(max_inter):
        prev_x=new_x
        gradient=derivative_func(prev_x)
        new_x=prev_x-step_multiplier * gradient

        step_size=abs(new_x-prev_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            print('times:', n)
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x,deriv_list= Grad_desc(dg,0.5, 0.02, 0.001,500)
print('local min occur at:',local_min )
print('Number of steps:', len(list_x))
```

```python
local_min, list_x,deriv_list= Grad_desc(dg,0.1, 0.02, 0.001,500)
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
value=np.array(list_x)
plt.scatter(list_x, g(value), color='r', s=100)
plt.plot(x_2, g(x_2))
plt.xlabel('x_2', fontsize=16)
plt.ylabel('g(x_2)', fontsize=16)
plt.title('cost_function', fontsize=16)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)

plt.subplot(1,2,2)
plt.grid()
plt.scatter(list_x, deriv_list, color='red', s=100)
plt.plot(x_2, dg(x_2), linewidth=5, alpha=0.6)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.xlabel('x_2', fontsize=16)
plt.ylabel('dg(x_2)', fontsize=16)
plt.title(' Slope_cost_function', fontsize=16)
plt.show()
```

```python
def fpx(x,y):
    r=3**(-x**2 - y**2)
    return 2*x*log(3)*r/(r+1)**2


def fpy(x,y):
    r=3**(-x**2-y**2)
    return 2*y*log(3)*r /(r+1)**2
```

```python
multiplier=0.1
max_iter=200
params=np.array([1.8,1.0])

for n in range(max_iter):
    gradient_x=fpx(params[0],params[1])
    gradient_y=fpy(params[0],params[1])
    gradient=np.array([gradient_x,gradient_y])
    params=params-multiplier*gradient
    
print('values in gradient Array:', gradient)
print('mininum occurs at x value of:', params[0])
print('mininum occurs at y value of:', params[1])
print('The cos is:', f(params[0],params[1]))
```

# 3D plot for MSE cost Function

## make data for thetas


```python
nr_theta=5
thet_0=np.linspace(-1,3,nr_theta)
thet_1=np.linspace(-1,3,nr_theta)
plot_t0, plot_t1=np.meshgrid(thet_0, thet_1)
print(thet_0.shape) #1 dimension compared to what the meshgrid did to the later as seen below
print(plot_t0.shape)# 2D
```

## Cost MSE using nested for loop

```python
plot_cost=np.zeros((nr_theta,nr_theta))
print(plot_cost)

for i in range(nr_theta):
    for j in range(nr_theta):
        y_hat=plot_t0[i][j] + plot_t1[i][j] * x_5
        plot_cost[i][j]=mse(y_5, y_hat)
        
print(plot_cost)
print('shape of plot_t0', plot_t0.shape)
print('shape of plot_t1', plot_t1.shape)
print('shape of plot_cost', plot_cost.shape)
        
```

```python
# x values, y values , array of theta parameters(theta0 at index 0 and theta1 at 1)

def grad(x,y,thetas):
    n=y.size
    
    theta0_slope=(-2/n) * sum(y-thetas[0]-thetas[1]*x)
    theta1_slope=(-2/n) * sum((y-thetas[0]-thetas[1]*x)*x)
    
    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multipier=0.01
thetas=np.array([2.9,2.9])

for i in range(1000):
    thetas= thetas- multiplier * grad(x_5, y_5, thetas)
    
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print("MSE is:", mse(y_5, thetas[0] + thetas[1] * x_5))
```
